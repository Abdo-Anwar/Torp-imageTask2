```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def display_img(img):
    new_img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    plt.imshow(new_img)
    plt.show()

img=cv2.imread("Squirrel.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.subplot(221)
plt.imshow(img,cmap="gray")
plt.title('Original')
plt.axis('off')


blur=cv2.GaussianBlur(img,(13,13),100)
edges=cv2.subtract(img,blur)

plt.subplot(222)
plt.imshow(edges,cmap="gray")
plt.title('Edges')
plt.axis('off')


img2=cv2.imread("Squirrel.png")
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
G_blur=cv2.GaussianBlur(gray,(3,3),100)
Sobel_X=cv2.Sobel(G_blur,-1,1,0) 
Sobel_Y=cv2.Sobel(G_blur,-1,0,1)
Sobel_or=cv2.bitwise_or(Sobel_X,Sobel_Y)

plt.subplot(223)
plt.imshow(Sobel_or,cmap="gray")
plt.title('Sobel')
plt.axis('off')


plt.show()
```
