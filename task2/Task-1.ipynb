```python
import cv2
import matplotlib.pyplot as plt
import numpy as np



img=cv2.imread("Squirrel.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.subplot(221)
plt.imshow(img,cmap="gray")
plt.title('Original')
plt.axis('off')


blur=cv2.GaussianBlur(img,(13,13),100)
edges=cv2.subtract(img,blur)

plt.subplot(222)
plt.imshow(edges,cmap="gray")
plt.title('Edges')
plt.axis('off')



plt.show()
```
